# ICA on log returns
Apply PCA on log returns of stocks and use pyfolio for analysis

```python
import yfinance as yf
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
from matplotlib import pyplot as plt

data = yf.download("GOOGL AAPL AMD TSLA NFLX TXN AMAT AMZN INTC NVDA ADBE CSCO V IBM QCOM MSFT XLNX HPQ VZ T", start="2011-05-21", end="2021-05-20")
tech = data[["Adj Close","Volume"]]
# data = yf.download("BAC JPM GS C MS HSBC BNPQY BCS LYG UBS CS WFC RY SCHW TD HDB USB PNC TFC BNS", start="2011-05-21", end="2021-05-20")
# fin = data[["Adj Close","Volume"]]
```

```python
#returns
#returns tech
rs_tech = tech["Adj Close"].apply(np.log).diff(1) 
# rs_tech.plot(title='Daily Returns of the Stocks in the tech companies')

# #returns fin
# rs_fin = fin["Adj Close"].apply(np.log).diff(1) 
# # rs_fin.plot(title='Daily Returns of the Stocks in the finance companies')
```

```python
train_tech, test_tech = train_test_split(rs_tech, test_size=0.2, random_state=42, shuffle=False)
# train_fin, test_fin = train_test_split(rs_fin, test_size=0.2, random_state=42, shuffle=False)
```

```python
train_tech.fillna(0)
```

```python
from sklearn.decomposition import FastICA
ica = FastICA(n_components=10, random_state=0)
ica_tech = ica.fit(train_tech.fillna(0))
# ica_fin = ica.fit(train_fin.fillna(0))
```

```python
fig, ax = plt.subplots(1, figsize=(20, 10)) 
pan = yf.download("GOOGL AAPL AMD TSLA NFLX TXN AMAT AMZN INTC NVDA ADBE CSCO V IBM QCOM MSFT XLNX HPQ VZ T", start="2011-05-20", end="2019-05-21")
ret = pan["Adj Close"].apply(np.log).diff(1)
for i in range(10):
    ret.dot(ica.components_[i]).cumsum().plot(ax=ax,fontsize=14)
ax.legend([str(i) for i in range(1,11)],fontsize=14)
ax.set_title("Independent Components",fontsize=14)
ax.set_ylabel('Returns', fontsize=14)
ax.set_xlabel('Date', fontsize=14)
ax.grid()
```

```python
fig, axes = plt.subplots(1)
#tech
ica_tech = pd.Series(index=train_tech.columns, data=ica.components_[4])
ica_tech.plot(ax=axes, xticks=np.arange(0, 20, 1.0),  title='Dominant Independent Component',fontsize=14)
axes.set_ylabel('Returns', fontsize=14)
axes.set_xlabel('Date', fontsize=14)

# #fin
# ica_fin = pd.Series(index=train_fin.columns, data=ica_fin.components_[0])
# ica_fin.plot(ax=axes[1], xticks=np.arange(0, 20, 1.0),  title='First Independent Component of the finance sector')

fig.set_size_inches(18, 8) 
```

```python
ica_tech.sort_values()
```

```python
from matplotlib.pyplot import figure
fig = figure(num=None, figsize=(15, 8))
plt.bar(ica_tech.sort_values().index[-10:],ica_tech.sort_values().values[-10:])
plt.ylabel('Asset weights', fontsize=14)
plt.xlabel('Tickers', fontsize=14)
plt.tick_params(axis='x',labelsize=14)
plt.tick_params(axis='y',labelsize=14)
plt.title("Top 10 assets",fontsize=14)
```

```python
test_tech.dot(ica_tech)
```

```python
test_tech.transpose().mean()
```

```python
fig,ax=plt.subplots(1, figsize=(15, 8)) 
test_tech.dot(ica_tech).cumsum().plot(ax=ax, style='r-')
test_tech.transpose().mean().cumsum().plot(ax=ax, style='b-',fontsize=14)
ax.legend(['IC1 Portfolio returns','All Stocks returns'],fontsize=14)
ax.set_title("ICA Weighted Returns and Non-Weighted Returns",fontsize=14)
ax.set_ylabel('Returns', fontsize=14)
ax.set_xlabel('Date', fontsize=14)
ax.grid()
```

```python
import pyfolio as pf
print("Unweighted returns ICA")
pf.create_returns_tear_sheet(test_tech.transpose().mean())
```

```python
print("Weighted returns ICA")
pf.create_returns_tear_sheet(test_tech.dot(ica_tech))
```

## Top 10

```python
dominant_assets = ica_tech.sort_values().index[-10:]
dominant_assets = " ".join(dominant_assets)
dominant_assets
```

```python
test = yf.download(dominant_assets, start="2019-05-19", end="2021-05-20")
returns_test = test["Adj Close"].apply(np.log).diff(1)
returns_test = returns_test.stack()
returns_test.index = returns_test.index.set_names(['date', 'asset'])
returns_test = returns_test.unstack()
returns_test = returns_test.stack()
returns_test
```

```python
test["Adj Close"].apply(np.log).diff(1).mean(1).dropna()
```

```python
fig, ax = plt.subplots(1, figsize=(15, 8)) 
test["Adj Close"].apply(np.log).diff(1).mean(1).dropna().cumsum().plot(ax=ax, style='y-')
test_tech.dot(ica_tech).cumsum().plot(ax=ax, style='r-')
test_tech.mean(1).cumsum().plot(ax=ax, style='b-',fontsize=14)
ax.set_title("Top 10 ICA Stocks Returns, ICA Weighted Returns and Non-Weighted Returns",fontsize=14)
ax.legend(['Top 10 stocks portfolio returns','IC1 portfolio returns','All Stocks returns'],fontsize=14)
ax.set_ylabel('Returns', fontsize=14)
ax.set_xlabel('Date', fontsize=14)
ax.grid()
```

```python
mean_returns = []
for i in range(0,len(returns_test),10):
    mean_returns.append(returns_test[i:i+10].mean())
mean_returns
```

```python
df = pd.Series(mean_returns,index=sorted(list(set(returns_test.index.get_level_values('date')))))
```
